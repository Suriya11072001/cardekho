```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
df = pd.read_csv('preprocessed_car_data.csv')

# 1. Basic info
print("üîç Dataset Info:")
print(df.info())

# 2. Missing values
print("\n‚ùì Missing Value Count:")
print(df.isnull().sum())
#
# 3. Statistical Summary
print("\nüìä Summary Statistics:")
print(df.describe())

```

```python
#4. Univariate Plots

# Car Price distribution
# Load cleaned data
df = pd.read_csv('preprocessed_car_data.csv')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 4))
sns.histplot(df['Car_Price'], bins=50, kde=True)
plt.title('Distribution of Car Price')
plt.xlabel('Price (INR)')
plt.ylabel('Frequency')
plt.show()

```

```python
# Fuel Type count
plt.figure(figsize=(6, 4))
sns.countplot(x='Fuel_Type', data=df)
plt.title('Fuel Type Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
# Body Type
plt.figure(figsize=(10, 4))
sns.countplot(x='Body_Type', data=df, order=df['Body_Type'].value_counts().index)
plt.title('Body Type Count')
plt.xticks(rotation=45)
plt.show()
```

```python

# Transmission Type
plt.figure(figsize=(6, 4))
sns.countplot(x='Transmission_Type', data=df)
plt.title('Transmission Type Distribution')
plt.show()
```

```python
# city count
plt.figure(figsize=(6, 4))
sns.countplot(x='City', data=df)
plt.title('city Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Bivariate Relationships

# Car Price vs Kilometers
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Kilometers_Driven', y='Car_Price', hue='Fuel_Type', data=df)
plt.title('Price vs Kilometers Driven')
plt.xlabel('Kilometers Driven')
plt.ylabel('Car Price')
plt.show()
```

```python
# Car Price vs Engine_CC
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Engine_CC', y='Car_Price', hue='Transmission_Type', data=df)
plt.title('Price vs Engine CC')
plt.xlabel('Engine CC')
plt.ylabel('Car Price')
plt.show()
```

```python
# Car Price vs Mileage
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Mileage_kmpl', y='Car_Price', hue='Fuel_Type', data=df)
plt.title('Price vs Mileage (kmpl)')
plt.xlabel('Mileage (kmpl)')
plt.ylabel('Car Price')
plt.show()
```

```python
# 6. Correlation Heatmap (Numerical features)
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("üîó Correlation Heatmap")
plt.show()
```

```python
categorical_columns = ['Fuel_Type','Body_Type', 'Transmission_Type', 'City']

for col in categorical_columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(df[col], order=df[col].value_counts().index, palette="Set2")
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
import pandas as pd

# Load cleaned data
df = pd.read_csv('preprocessed_car_data.csv')

# Drop high-cardinality or text-heavy columns
df.drop(['Car_Model', 'Car_Link'], axis=1, inplace=True)

# Define categorical columns for one-hot encoding
categorical_cols = ['Fuel_Type', 'Transmission_Type', 'Body_Type', 'City', 'Manufactured_By']

# One-hot encode
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Convert boolean columns to 0/1 integers
bool_cols = df_encoded.select_dtypes(include='bool').columns
df_encoded[bool_cols] = df_encoded[bool_cols].astype(int)

# Save
df_encoded.to_csv('encoded_car_data.csv', index=False)
print("üíæ Encoded data saved with 0/1 binary columns in 'encoded_car_data.csv'")
```
